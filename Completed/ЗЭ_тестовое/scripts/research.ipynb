### EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ankets = pd.read_csv("./data/Anket.csv", sep=";")
ml = pd.read_csv("./data/ML.csv", sep=";")
payments = pd.read_csv("./data/Payments.csv", sep=";", encoding="utf-8")
```

```python
ankets.drop('Unnamed: 0', axis=1, inplace=True)
ankets.head()
```

```python
print(ankets.info())
print(payments.info())
```

```python
ankets.describe()
```

```python
payments.describe()
```

```python
num_ankets = ankets.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(8, 8))
sns.heatmap(num_ankets.corr(), annot=True, fmt='.2f')
plt.show()
```

```python
num_payments = payments.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(8, 8))
sns.heatmap(num_payments.corr(), annot=True, fmt='.2f')
plt.show()
```

```python
ankets_without_drafts = ankets.drop('DraftId', axis=1)
encoded_df = pd.get_dummies(ankets_without_drafts, drop_first=True)
corr_matrix_all = encoded_df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix_all, annot=False, cmap='coolwarm')
plt.title("Корреляционная матрица всех признаков (включая категориальные)")
plt.show()
```

```python
payments_without_drafts = ankets.drop('DraftId', axis=1)
encoded_df = pd.get_dummies(payments_without_drafts, drop_first=True)
corr_matrix_all = encoded_df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix_all, annot=False, cmap='coolwarm')
plt.title("Корреляционная матрица всех признаков")
plt.show()
```

```python
full_df = pd.merge(payments, ankets, on="DraftId", how="left")
full_df.shape
```

```python
amount_by_gender = full_df.groupby('Gender')['Amount'].sum().reset_index()
```

Известно, что преимущественно обращаются за кредитом мужчины -- из задания, но судя по графику сумма выданных кредитов по полу не сильно различается;
Но видимо это пояснение чтобы понять что мужчины с цифоркой 1 в гендере ;)

```python
plt.figure(figsize=(5, 5))
sns.barplot(x='Gender', y='Amount', data=amount_by_gender)

plt.title('Сумма выданного займа в зависимости от пола')
plt.xlabel('Пол')
plt.ylabel('Сумма выданного займа')
plt.xticks(rotation=0)

plt.show()
```
